```python
from langchain_core.prompts import ChatPromptTemplate
from langchain_openai import ChatOpenAI
```

```python
template = ChatPromptTemplate.from_messages([
    ("system", "You are a helpful assistant that translates English to persian."),
    ("user", "Translate the following English text to persian: {text}"),
])

chat = ChatOpenAI()
formatted_message = template.format_messages(text="I love programming.")
response = chat.invoke(formatted_message)
print(response)
```

```python
template = ChatPromptTemplate.from_messages([
    ("system", "You are a helpful assistant that translates English to persian."),
    ("user", "Translate the following English text to persian: {text}"),
])

ollama = ChatOllama(model="llama2")
formatted_message = template.format_messages(text="I love programming.")
response = ollama.invoke(formatted_message)
print(response)
```

```python
ollama = ChatOllama(model="llama3.2")

messages = [
    (
        "system",
        "You are a helpful assistant that translates English to Persian. Translate the user sentence.",
    ),
    ("human", "I love programming."),
]
ai_msg = ollama.invoke(messages)
print(ai_msg.content)
```
